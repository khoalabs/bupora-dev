##### convert __indent__

```python
# @title ##### 🔄 convert __indent__
import re, subprocess, ctypes, ctypes.wintypes

if ctypes.windll.user32.OpenClipboard(None):
  CF_UNICODETEXT = 13
  handle  = ctypes.windll.user32.GetClipboardData
  handle.argtypes = (ctypes.wintypes.UINT,)
  handle.restype = ctypes.wintypes.HANDLE
  handle  =  handle(CF_UNICODETEXT)
  code  = ctypes.windll.kernel32.GlobalLock
  code.argtypes = (ctypes.wintypes.HGLOBAL,)
  code.restype = ctypes.wintypes.LPVOID
  code = ctypes.wstring_at(code(handle))
  unlock  = ctypes.windll.kernel32.GlobalUnlock
  unlock.argtypes = (ctypes.wintypes.HGLOBAL,)
  unlock.restype = ctypes.wintypes.BOOL
  unlock(handle)
  ctypes.windll.user32.CloseClipboard()
  code = re.sub(r'^( +)', lambda m: ' ' * ((len(m.group(1)) // 4) * 2), code, flags=re.M)
  subprocess.run(['clip.exe'], input=code.strip().encode('utf-16le'), check=True)
```

##### convert __SVG__ to __optimization__

```python
import os, re, xml.etree.ElementTree

xml.etree.ElementTree.register_namespace('', 'http://www.w3.org/2000/svg')
paths = [
  r'D:\PY\site\assets\site\logo.svg',
]
paths = [x for x in paths if os.path.isfile(x) and x.endswith('.svg')]
for i, p in enumerate(paths, start=1):
  tree = xml.etree.ElementTree.parse(p)
  root = tree.getroot()
  for attr in ['id', 'width', 'height']:
    root.attrib.pop(attr, None)
  if 'viewBox' not in root.attrib:
    root.set('viewBox', '0 0 24 24')
  root.set('fill', 'currentColor')
  for elem in root.iter():
    for attr in ['id', 'style', 'stroke-width']:
      elem.attrib.pop(attr, None)
    if elem is not root:
      elem.attrib.pop('fill', None)
    if elem.tag.endswith('path') and 'd' in elem.attrib:
      elem.attrib['d'] = re.sub(r'\s+', ' ', elem.attrib['d'].strip())
  svg_str = xml.etree.ElementTree.tostring(root, encoding='unicode')
  svg_str = re.sub(r',', ' ', svg_str)
  svg_str = re.sub(r'\s+', ' ', svg_str)
  svg_str = re.sub(r'>\s+<', '><', svg_str)
  svg_str = re.sub(r'\s+xmlns="[^"]+"', '', svg_str)
  svg_str = re.sub(r'\s+/>', '/>', svg_str)
  svg_str = svg_str.strip()
  _p = os.path.join(os.path.dirname(p), '_' + os.path.basename(p))
  with open(_p, 'w', encoding='utf-8') as f:
    f.write(svg_str)
  os.remove(p)
  os.rename(_p, p)
  print(f'✅ {i}/{len(paths)}: {p}')
```
